Solve the proportion:

$$
\frac{8}{36} = \frac{10}{n}
$$

solution process

$$
\frac{a}{b} = \frac{c}{d}
$$
$$
a \cdot d = b \cdot c
$$

```python
left_number = 8
left_denominator = 36
right_nummerator = 10

n = (left_denominator * right_nummerator) / left_number
print(n)
```

```python
%reset -f

def solve_proportion(a, b, c):
    # a / b = c / d
    # a * d = b * c
    d = (b * c) / a
    return d

result = solve_proportion(8, 36, 10)
print(result)
```
