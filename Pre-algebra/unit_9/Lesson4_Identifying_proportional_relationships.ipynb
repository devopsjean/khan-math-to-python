### Does the following table show a proportional relationship between the variables *x* and *y*?

| x      | y  |
|--------|----|
| 5/6    | 2  |
| 25/6   | 10 |
| 125/6  | 50 |

**Choose 1 answer:**

A. Yes  
B. No

```python
x = [5/6, 25/6, 125/6]
y = [2, 10, 50]

ratio = y[0] / x[0]

is_proportional = True

for i in range(len(x)):
    if y[i] / x[i] != ratio:
        is_prortaion= False
        break

print(is_proportional)
```
