Calvin cleans 3/5 of his bathroom with 1/3 of a bottle of cleaning solution.
At this rate, what fraction of the bottle of cleaning solution will Calvin use to clean his entire bathroom?

tip

bottle per bathroom = bottle / bathroom = (1/3) / (3/5)


```python
from fractions import Fraction

bathroom_cleaned = Fraction(3, 5)
solution_used = Fraction(1, 3)

rate = solution_used / bathroom_cleaned

total_solution_needed = rate * 1

print(total_solution_needed)
```

```python
from fractions import Fraction as F

result = F(1,3) / F(3,5)

print(result)
```
