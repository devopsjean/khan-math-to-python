The cake store is having a 25% off sale on all of its cakes.
If the cake you want regularly cost $9, how much would you save with the discount?

```python
regular_costs = float(input('Enter regular cost ($): '))
discount = float(input('What percent for discount (%): '))

print(regular_costs * (discount / 100))
```

The manager at Gabriela's Furniture Store is trying to figure out how much to charge for a book shelf that just arrived. The book shelf was bought at  wholesale price of $147.00, and Gabriela's Furniture Sotre marks up all furniture by 60%.

At what price should the manager sell the book shelf

```python
wholesale_price = float(input('Enter wholesale price ($): '))
marksup_by = float(input('What percent for markup (%): '))

print(wholesale_price + (wholesale_price * (marksup_by / 100)))
```

William works at a nearby electronics store. He makes a commission of 6% on everything he slees. If he sells a computer for $764.00, how much money does William make in commision?

Whole sales amount x commission ratio(/100) = William's commission 

```python
sales_amount = float(input('Enter wholesale amount ($): '))
commission_rate = float(input('What is the commission rate (%): '))

total_commission = sales_amount * (commission_rate / 100)

print(f"${total_commission:.2f}")
```

Allam just finished a great meal at a restaurant in Wisconsin. The sales tax in Wisonsin is 5%, and it is customary to leave a tip of 15% for the server. The tip amount is calculated on the price of the meal before the tax is applied. (Sales tax is not caculated on tips.)

If Allam ordered a $15 meal, how much money will he have to pay with tax and tip?


```python
state_tax = float(input('Enter state tax rate (%): '))
tax_for_server = float(input('Enter tax for server ($): ')) 
meal_price = float(input('Enter meal price ($): '))

total_price = meal_price + (((state_tax + tax_for_server) / 100) * meal_price)

print(f"${total_price:.2f}")
```
