The cake store is having a 25% off sale on all of its cakes.
If the cake you want regularly cost $9, how much would you save with the discount?

```python
def discount_savings(regular_cost: float, discount_percent: float) -> float:
    return regular_cost * (discount_percent / 100)

# Fixed example from the prompt
example_saving = discount_savings(9.0, 25.0)
print(f"Example saving: ${example_saving:.2f}")

# Optional interactive run (keep False for reproducible GitHub/CI runs)
use_custom_input = False
if use_custom_input:
    regular_cost = float(input('Enter regular cost ($): '))
    discount = float(input('Enter discount percent (%): '))
    print(f"Custom saving: ${discount_savings(regular_cost, discount):.2f}")
```

The manager at Gabriela's Furniture Store is trying to figure out how much to charge for a book shelf that just arrived. The book shelf was bought at  wholesale price of $147.00, and Gabriela's Furniture Sotre marks up all furniture by 60%.

At what price should the manager sell the book shelf

```python
def sale_price_with_markup(wholesale_price: float, markup_percent: float) -> float:
    return wholesale_price * (1 + markup_percent / 100)

# Fixed example from the prompt
example_sale_price = sale_price_with_markup(147.0, 60.0)
print(f"Example sale price: ${example_sale_price:.2f}")

# Optional interactive run
use_custom_input = False
if use_custom_input:
    wholesale_price = float(input('Enter wholesale price ($): '))
    markup_percent = float(input('Enter markup percent (%): '))
    print(f"Custom sale price: ${sale_price_with_markup(wholesale_price, markup_percent):.2f}")
```

William works at a nearby electronics store. He makes a commission of 6% on everything he slees. If he sells a computer for $764.00, how much money does William make in commision?

Whole sales amount x commission ratio(/100) = William's commission 

```python
def commission_earned(sales_amount: float, commission_percent: float) -> float:
    return sales_amount * (commission_percent / 100)

# Fixed example from the prompt
example_commission = commission_earned(764.0, 6.0)
print(f"Example commission: ${example_commission:.2f}")

# Optional interactive run
use_custom_input = False
if use_custom_input:
    sales_amount = float(input('Enter sales amount ($): '))
    commission_percent = float(input('Enter commission percent (%): '))
    print(f"Custom commission: ${commission_earned(sales_amount, commission_percent):.2f}")
```

Allam just finished a great meal at a restaurant in Wisconsin. The sales tax in Wisonsin is 5%, and it is customary to leave a tip of 15% for the server. The tip amount is calculated on the price of the meal before the tax is applied. (Sales tax is not caculated on tips.)

If Allam ordered a $15 meal, how much money will he have to pay with tax and tip?


```python
def total_with_tax_and_tip(meal_price: float, tax_percent: float, tip_percent: float) -> float:
    # Tip is calculated on the meal price before tax.
    return meal_price + (meal_price * tax_percent / 100) + (meal_price * tip_percent / 100)

# Fixed example from the prompt
example_total = total_with_tax_and_tip(15.0, 5.0, 15.0)
print(f"Example total payment: ${example_total:.2f}")

# Optional interactive run
use_custom_input = False
if use_custom_input:
    meal_price = float(input('Enter meal price ($): '))
    tax_percent = float(input('Enter state tax percent (%): '))
    tip_percent = float(input('Enter tip percent (%): '))
    print(f"Custom total payment: ${total_with_tax_and_tip(meal_price, tax_percent, tip_percent):.2f}")
```
