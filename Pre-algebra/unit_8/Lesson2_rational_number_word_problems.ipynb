## Average Daily Balance Problem

The **average daily balance** is the mean of the balance in an account  
at the end of each day in a month.

The following table gives the **dates and amounts of the transactions**  
in Elliott's account in **June**.

| Day of June | Transaction type | Transaction amount (in dollars) |
|-------------|------------------|----------------------------------|
| 1  | Starting balance | 1223 |
| 10 | Deposit          | 615  |
| 15 | Withdrawal       | -63  |
| 22 | Withdrawal       | -120 |

There are **30 days in June**.

### Question

What is the **average daily balance** of Elliott's account for the month of **June**?

```python
start_balance = 1223                 # Starting balance on June 1

balance_1 = start_balance            # Balance from June 1 to June 9
days_1 = 10 - 1                      # Number of days this balance remains (9 days)

balance_2 = balance_1 + 615          # Balance after deposit on June 10
days_2 = 15 - 10                     # Days until the next transaction (5 days)

balance_3 = balance_2 - 63           # Balance after withdrawal on June 15
days_3 = 22 - 15                     # Days until the next transaction (7 days)

balance_4 = balance_3 - 120          # Balance after withdrawal on June 22
days_4 = 31 - 22                     # Days until the end of the month (9 days)

# Compute the weighted sum of daily balances
sum_of_daily_balances = (
    balance_1 * days_1 +
    balance_2 * days_2 +
    balance_3 * days_3 +
    balance_4 * days_4
)

# Calculate the average daily balance for June
average_daily_balance = sum_of_daily_balances / 30

print("balance_1:", balance_1, "days:", days_1)
print("balance_2:", balance_2, "days:", days_2)
print("balance_3:", balance_3, "days:", days_3)
print("balance_4:", balance_4, "days:", days_4)
print("sum_of_daily_balances:", sum_of_daily_balances)
print("average_daily_balance:", average_daily_balance)
```
