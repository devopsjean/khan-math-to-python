Fathi has $1.10 in his printing account. Each sheet of paper he uses reduces his printing account balance by $0.25. Fathi wats to print out a PDF document that is 47 pages long. To save paper, he decides to print on both sides of each sheet and to print two pages on each side of the sheet.

After Fathi prints, what will be the balance in his printing account.

1.10(balance) - [((47(total)pages) / 2(both sides of each sheet) / 2(topages on each side of the sheet)) x 0.25(cost_per_sheet)] = totla_left_balance 

```python
balance = 1.10
total_pages = 47
cost_per_sheet = 0.25

pages_per_sheet = 4

sheets_needed = (total_pages + pages_per_sheet -1) // pages_per_sheet
# ceil(total_pages /pages_per_sheet) = (total_pages + pages_per_sheet -1) // pages_per_sheet 

total_cost = sheets_needed * cost_per_sheet
left_balance = balance - total_cost 

print(f"sheet_needed: {sheets_needed} sheets")
print(f"total_cost: ${total_cost}")
print(f"left_balance: ${left_balance}")
```
