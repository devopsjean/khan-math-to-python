Ashely recently opened a store that uses only natural ingredients. She wants to advertise her products by distributing bas of samples in her neighborhood. It takes one person 2/3 of a minute to prepare on bag.

How many hours will ti take Ashley and 4 of her griends to prepare 1575 bags of samples?

taking_time_per_bag = 2/3 of a minute = 40s

```python
def prep_hours(total_bags: float, workers: float, minutes_per_bag: float) -> float:
    total_minutes = total_bags * minutes_per_bag
    minutes_per_worker = total_minutes / workers
    return minutes_per_worker / 60

# Fixed example from the prompt
example_hours = prep_hours(total_bags=1575, workers=5, minutes_per_bag=2/3)
print(f"Example hours needed: {example_hours:.2f}")

# Optional interactive run
use_custom_input = False
if use_custom_input:
    total_bags = float(input('Enter number of bags: '))
    workers = float(input('Enter number of workers: '))
    minutes_per_bag = float(input('Enter minutes per bag: '))
    print(f"Custom hours needed: {prep_hours(total_bags, workers, minutes_per_bag):.2f}")
```
